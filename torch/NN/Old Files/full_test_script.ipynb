```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
device = torch.device('cpu') # change to cuda
```

```python
tensor_transform = transforms.ToTensor()

dataset = datasets.MNIST(root='./data', 
                         train = True, 
                         download = True, 
                         transform = tensor_transform)

train_set, test_set = torch.utils.data.random_split(dataset, [50000, 10000])

loader = torch.utils.data.DataLoader(dataset = train_set,
                                    batch_size = 32,
                                    shuffle = True,
                                    generator=torch.Generator(device=device))
```

```python
dataiter = iter(loader)
images, labels = next(dataiter)
print(torch.min(images), torch.max(images))
```

```python
from AutoEncoderCNN import AE_CNN
model = AE_CNN().to(device)

loss_function = torch.nn.BCELoss()

optimizer = torch.optim.Adam(model.parameters(), lr = 0.01, weight_decay=1e-5)
```

```python
model
```

```python
epochs = 1
outputs = []

for epoch in range(epochs):
    for (image, _) in loader:
        image = image.to(device)
        #image = image.reshape(-1, 28*28) # no reshape for CNN
        
        recon = model(image)
        loss = loss_function(recon, image)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Epoch: {epoch + 1}, Loss: {loss.item():.4f}')
    outputs.append((epoch, image, recon))
```

```python
encoder = model.encoder
decoder = model.decoder
```

```python
auto_encoder_output = []
encoded_imgs_arr = []
decoded_imgs_arr = []
for x in range(len(test_set)):
    with torch.no_grad():
        inp = test_set.__getitem__(x)[0].to(device)#.flatten()
        
        encoded_imgs = encoder(inp)
        encoded_imgs_arr.append(encoded_imgs)
        
        decoded_imgs = decoder(encoded_imgs)
        decoded_imgs_arr.append((inp, decoded_imgs))
        
        auto_encoder_output.append((inp, model(inp)))
```

```python
amt = 10
plt.figure(figsize=(20,4))


for k in range(amt):
    ax = plt.subplot(4, amt, k+1)
    
    img = decoded_imgs_arr[k][0].detach().cpu().numpy()
    #img = img.reshape(-1,28,28)
    plt.imshow(img[0])
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    ax = plt.subplot(4,amt, k+1+amt)

#     img = encoded_imgs_arr[k].detach().cpu().numpy()
#     #img = img.reshape(-1,4,8)
#     plt.imshow(img[0])
#     ax.get_xaxis().set_visible(False)
#     ax.get_yaxis().set_visible(False)
    
    ax = plt.subplot(4,amt, k+1+2*amt)

    recon = decoded_imgs_arr[k][1].detach().cpu().numpy()
    recon = recon.reshape(-1,28,28)
    plt.imshow(recon[0])
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    ax = plt.subplot(4,amt, k+1+3*amt)

    recon = auto_encoder_output[k][1].detach().cpu().numpy()
    recon = recon.reshape(-1,28,28)
    plt.imshow(recon[0])
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```
